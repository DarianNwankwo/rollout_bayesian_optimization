```julia
# Global parameters
MAX_SGD_ITERS = 100
BATCH_SIZE = 5
HORIZON = 0
MC_SAMPLES = 500
BUDGET = 1;
```

```julia
# Setup toy problem
testfn = TestFunction(
    1, [0. 1.], [.5],
    x -> 0.,
    ∇x -> [0.]
)
lbs, ubs = testfn.bounds[:,1], testfn.bounds[:,2]

# Setup low discrepancy random number stream
lds_rns = gen_low_discrepancy_sequence(MC_SAMPLES, testfn.dim, HORIZON+1);
rns = randn(MC_SAMPLES, testfn.dim+1, HORIZON+1);
```

```julia
# Gather initial samples/experimental data
N, θ = 1, [.25]
X = [.15, .85]
X = reshape(X, 1, length(X))
ψ = kernel_matern52(θ);
sur = fit_surrogate(ψ, X, testfn.f);
```

```julia
res = optimize_hypers_optim(sur, kernel_matern52)
```

```julia
θopt = Optim.minimizer(res)
scaled_ψ = kernel_scale(kernel_matern52, θopt)
optsur = fit_surrogate(scaled_ψ, X, testfn.f);
```

```julia
domain = filter(x -> !(x in X), lbs[1]:.01:ubs[1])
scatter(sur.X', sur.y)
plot!(domain, [sur([x]).μ for x in domain],
    ribbons=2*[sur([x]).σ for x in domain], label="μ ± 2σ")
plot!(domain, [sur([x]).EI for x in domain], label="EI")
```

```julia
domain = filter(x -> !(x in X), lbs[1]:.01:ubs[1])
scatter(optsur.X', sur.y)
plot!(domain, [optsur([x]).μ for x in domain],
    ribbons=2*[optsur([x]).σ for x in domain], label="μ ± 2σ")
plot!(domain, [optsur([x]).EI for x in domain], label="EI",
    ylim=(-2, 2))
```

### Investigations on Trivial Horizon Acquisition Function Gradient Computations
For $h \geq 1$ our gradient estimates have the correct sign, but for $h = 0$, our sign is incorrect. I suspect this is due to the nature of how we differentiate trajectories for non-trivial horizons. For trivial horizons, the gradient of the acquisition function is something we have a handle on. Let's compare the thing we have a handle on against how we compute trajectories.

```julia
eis, ∇eis, eis_other, ∇eis_other = [], [], [], [] # eis are rollout
for random_number_stream in [rns, lds_rns]
    rollout_ei = [0 0] # Sample mean and variance tuples
    ∇rollout_ei = [0 0] # Sample mean and variance tuples
    
    println("Total '|'s => $(length(domain))")
    for x0 in domain
        print("|$x0")
        # Grab each input location and convert to a column vector
        x0 = [x0]

        αxs, ∇αxs, αxs_other, ∇αxs_other = [], [], [], []
        # Monte-carlo integrate trajectory for x0
        for sample in 1:MC_SAMPLES
            # Make a copy of our surrogate to pass to the trajectory struct
            # for fantasized computations
            fsur = Base.deepcopy(sur)
            fantasy_ndx = size(fsur.X, 2) + 1

            # Rollout trajectory
            T = Trajectory(fsur, x0, fantasy_ndx; h=HORIZON)
            rollout!(T, lbs, ubs; rnstream=random_number_stream[sample,:,:])
            
            # Evaluate rolled out trajectory
            push!(αxs, α(T))
            push!(∇αxs, first(∇α(T)))
        end # endfor sample
        
        # Compute function and gradient analytically
        fsur2 = Base.deepcopy(sur)
        fsx = fsur2(x0)

        # Evaluate analytic ∇EI
        αxs_other = first(fsx.EI)
        ∇αxs_other = first(fsx.∇EI)
        push!(eis_other, αxs_other)
        push!(∇eis_other, ∇αxs_other)

        # Average trajectories
        μx = sum(αxs) / length(αxs)
        ∇μx = sum(∇αxs) / length(αxs)
        σx = sum((αxs .- μx) .^ 2) / (MC_SAMPLES-1)
        ∇σx = sum((∇αxs .- ∇μx) .^ 2) / (MC_SAMPLES-1)

        # Update history
        # sx = sur(x0)
        # μx += ei(sx.μ, sx.σ, minimum(sur.y)) + poi(sx.μ, sx.σ, minimum(sur.y))
        rollout_ei = vcat(rollout_ei, [μx σx])
        ∇rollout_ei = vcat(∇rollout_ei, [∇μx ∇σx])
    end # endfor x0
    rollout_ei = rollout_ei[2:end, :]
    ∇rollout_ei = ∇rollout_ei[2:end, :];
    
    push!(eis, rollout_ei)
    push!(∇eis, ∇rollout_ei)
end
```
